```python
from tkinter import *
```

```python
frame1 = Frame(root)
frame1.pack()
titleLabel = Label(frame1 , text="Tic Tac Toe" , font=("Arial" , 26) , bg="orange" , width=16 )
titleLabel.grid(row=0 , column=0)

optionFrame = Frame(root , bg="grey")
optionFrame.pack()

frame2 = Frame(root , bg="yellow")
frame2.pack()

board = { 1:" " , 2:" " , 3:" ",
          4:" " , 5:" " , 6:" ",
          7:" " , 8:" " , 9:" " }

turn = "x"
game_end = False
mode = "singlePlayer"
```

```python
def changeModeToSinglePlayer(): 
    global mode 
    mode = "singlePlayer"
    singlePlayerButton["bg"] = "lightgrey"
    multiPlayerButton["bg"] = "lightgrey"
```

```python
def changeModeToMultiplayer():
    global mode 
    mode = "multiPlayer"
    multiPlayerButton["bg"] = "lightgreen"
    singlePlayerButton["bg"] = "lightgrey"
```

```python
def updateBoard():
    for key in board.keys():
        buttons[key-1]["text"] = board[key]
```

```python
def checkForWin(player):
    # rows
    if board[1] == board[2] and board[2] == board[3] and board[3] == player:
        return True
    
    # It checks if the value of the variable "board" at index 1 is equal to the value of the variable "board" at index 2 and 
#     if the value of the variable "board" at index 2 is equal to the value of the variable "board" at index 3 and if the value
#     of the variable "board" at index 3 is equal to the value of the variable "player". If all these conditions are true, then
#     the function will return True. "board" is a variable that likely represents a game board or grid in a game, and "player" is
#     a variable that represents a player in the game. The double equals sign "==" is a comparison operator used to check if two 
#     values are equal.
    
    elif board[4] == board[5] and board[5] == board[6] and board[6] == player:
        return True
    
    elif board[7] == board[8] and board[8] == board[9] and board[9] == player:
        return True

    # columns
    elif board[1] == board[4] and board[4] == board[7] and board[7] == player:
        return True
    
    elif board[2] == board[5] and board[5] == board[8] and board[8] == player:
        return True
    
    elif board[3] == board[6] and board[6] == board[9] and board[9] == player:
        return True
    
    # diagonals
    elif board[1] == board[5] and board[5] == board[9] and board[9] == player:
        return True
    
    elif board[3] == board[5] and board[5] == board[7] and board[7] == player:
        return True
    

    return False
```

```python
def checkForDraw():
    for i in board.keys():
        if board[i] == " ":
            return False
    
    return True

# This code defines a function named `checkForDraw()` which takes no arguments. 

# The function iterates through all the keys in the `board` dictionary using a for loop. 

# For each key, the function checks if the corresponding value in the dictionary is equal to a space character (" "). 

# If the value is a space character, it means that the game is not yet over and there are still empty spaces on the board. 
# In this case, the function returns `False`.

# If all the values in the `board` dictionary are not equal to a space character, it means that the game has ended in a draw.
# In this case, the function returns `True`.
```

```python
def restartGame():
    global game_end
    game_end = False
    for button in buttons:
        button["text"] = " "

    for i in board.keys():
        board[i] = " "

    titleLabel = Label(frame1 , text="Tic Tac Toe" , font=("Arial" , 30) , bg="orange" , width=15 )
    titleLabel.grid(row=0 , column=0)
    
    
# This is a Python function called `restartGame()` that resets a Tic Tac Toe game. 

# The line `global game_end` allows the function to access and modify the global variable `game_end`.

# The `for` loop sets the text of each button in the game board to an empty string.

# The second `for` loop resets the values of each cell in the game board dictionary to an empty string.

# The remaining lines create a new `Label` widget with the title "Tic Tac Toe" and place it in the first row and column of a frame. The widget has a font size of 30, an orange background, and a width of 15.
```

```python
def minimax(board , isMaximizing):
    
    if checkForWin("o"):
        return 1 
    
    if checkForWin("x"):
        return -1
    
    if checkForDraw():
        return 0
    
    if isMaximizing:
        bestScore = -100

        for key in board.keys():
            if board[key] == " ":
                board[key] = "o"
                score = minimax(board , False) # minimax
                board[key] = " "
                if score > bestScore : 
                    bestScore = score 
        
        return bestScore
    
    else:
        bestScore = 100

        for key in board.keys():
            if board[key] == " ":
                board[key] = "x"
                score = minimax(board , True) # minimax
                board[key] = " "
                if score < bestScore : 
                    bestScore = score 
        
        return bestScore

# This is a Python function called `minimax` that implements the minimax algorithm to determine the best move for a Tic Tac Toe
# game. 

# The function takes two arguments: `board`, which is a dictionary representing the current state of the game board, and
#     `isMaximizing`, a boolean value indicating whether the function is called for the maximizing player (computer) or 
# the minimizing player (human).

# The first three `if` statements check for win, loss, or draw conditions and return 1, -1, or 0 respectively.

# If `isMaximizing` is `True`, the function tries to find the best move for the maximizing player by assigning a score to each
# possible move. It initializes `bestScore` to -100 and loops through each empty cell in the board. For each empty cell, 
# it places an "o" and recursively calls `minimax` with `isMaximizing` set to `False` to simulate the minimizing player's turn.
# It then restores the original board state and updates `bestScore` if the current score is higher than the current best score.
# Finally, it returns `bestScore`.

# If `isMaximizing` is `False`, the function tries to find the best move for the minimizing player by assigning a score to
# each possible move. It initializes `bestScore` to 100 and loops through each empty cell in the board. For each empty cell, 
# it places an "x" and recursively calls `minimax` with `isMaximizing` set to `True` to simulate the maximizing player's turn.
# It then restores the original board state and updates `bestScore` if the current score is lower than the current best score.
# Finally, it returns `bestScore`.
```

```python
def playComputer():
    bestScore = -100
    bestMove = 0

    for key in board.keys():
        if board[key] == " ":
            board[key] = "o"
            score = minimax(board , False) # minimax
            board[key] = " "
            if score > bestScore : 
                bestScore = score 
                bestMove = key

    board[bestMove] = "o"

    
# This is a Python function called `playComputer` that represents the computer's turn in a game of tic-tac-toe. 

# The function first initializes two variables: `bestScore` and `bestMove`. `bestScore` is set to -100, which is a lower bound
# for the scores that can be achieved using the minimax algorithm. `bestMove` is set to 0, which will be updated later to 
# represent the key of the best move.

# The function then iterates over each key in the `board` dictionary using a `for` loop. If the value associated with the
# current key is an empty space, indicated by `" "`, the code sets the value to `"o"`, which represents the computer's move at
# this stage of the game.

# The code then calls the `minimax` function with `board` and `isMaximizing` parameters updated for the next level of the game
# tree. The `isMaximizing` parameter is set to `False` to indicate that the next level of the game tree will be a minimizing 
# level.

# The resulting score from the recursive call to `minimax` is stored in the `score` variable. The code then sets the value
# associated with the current key back to `" "`, effectively undoing the move that was just made.

# If the `score` is greater than the current `bestScore`, the code updates both `bestScore` and `bestMove`. 
# This ensures that `bestScore` always represents the highest score that can be achieved by the computer at this level of the
# game tree, and `bestMove` represents the key of the cell that leads to this score.

# Finally, the function sets the value associated with `bestMove` to `"o"`, which represents the computer's best move according
# to the minimax algorithm.
```

```python
def play(event):
    global turn,game_end
    if game_end:
        return
    
    button = event.widget
    buttonText = str(button)
    clicked = buttonText[-1]
    if clicked == "n" :
        clicked = 1
    else :
        clicked = int(clicked)
    
    if button["text"] == " ":
        if turn == "x" :
            board[clicked] = turn
            if checkForWin(turn):
                winningLabel = Label(frame1 , text=f"{turn} wins the game", bg="orange", font=("Arial" , 26),width=16 )
                winningLabel.grid(row = 0 , column=0 , columnspan=3)
                game_end = True

            turn = "o"

            updateBoard()

            if mode == "singlePlayer":

                playComputer()

                if checkForWin(turn):
                    winningLabel = Label(frame1 , text=f"{turn} wins the game", bg="orange", font=("Arial" , 26),width=16   )
                    winningLabel.grid(row = 0 , column=0 , columnspan=3)
                    game_end = True

                turn = "x"

                updateBoard()

           
            
        else:
            board[clicked] = turn
            updateBoard()
            if checkForWin(turn):
                winningLabel = Label(frame1 , text=f"{turn} wins the game" , bg="orange", font=("Arial" , 26),width=16)
                winningLabel.grid(row = 0 , column=0 , columnspan=3)
                game_end = True
            turn = "x"

        
        if checkForDraw():
            drawLabel = Label(frame1 , text=f"Game Draw" , bg="orange", font=("Arial" , 26), width = 16)
            drawLabel.grid(row = 0 , column=0 , columnspan=3)


            
# This is a Python function called `play` that handles the player's turn in a game of tic-tac-toe. 

# The function first declares two global variables: `turn` and `game_end`. `turn` represents whose turn it is currently,
# either "x" or "o". `game_end` is a boolean that indicates whether the game has ended or not.

# The function then checks if `game_end` is True, which means the game has ended. If so, the function returns without doing 
# anything else.

# The function then gets the button that was clicked by the player from the `event` parameter, which is passed as an argument
# when the function is called. The function extracts the index of the button from its name, which is stored in the `buttonText`
# variable.

# If the index is equal to "n", it means that the button is the center button, so the function sets `clicked` to 1. Otherwise,
# it converts the index to an integer and stores it in `clicked`.

# If the text displayed on the button is an empty space, indicated by `" "`, the function updates the `board` dictionary with
# the current player's move at the cell represented by `clicked`. It then calls the `checkForWin` function to check if
# the current player has won the game. If so, it displays a winning message and sets `game_end` to True.

# The function then updates `turn` to be the next player's turn and calls `updateBoard` to update the game board display.

# If the game mode is "singlePlayer", the function calls `playComputer` to simulate the computer's turn. It then checks if 
# the computer has won and displays a message if so. It updates `turn` to be the next player's turn and calls `updateBoard` again.

# If the current player's move did not result in a win and all cells are occupied, the function calls `checkForDraw` to check 
# if the game has ended in a draw. If so, it displays a message indicating a draw.


# Function to play
```

```python
# Tic Tac Toe Board 

#  First row 

button1 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30) , bg="yellow" , relief=RAISED , borderwidth=5)
button1.grid(row = 0 , column=0)
button1.bind("<Button-1>" , play)

# This is a Python code snippet that creates a button widget, `button1`, and adds it to a frame called `frame2`.
# The `Button` function creates a button widget with the specified parameters. The `text` parameter sets the text displayed on
# the button. In this case, the text is an empty space, indicated by `" "`. 
# The `width` and `height` parameters set the dimensions of the button in pixels. 
# The `font` parameter sets the font used for the text on the button. 
# The `bg` parameter sets the background color of the button. 
# The `relief` parameter sets the appearance of the button border.
# The `borderwidth` parameter sets the width of the button border in pixels.
# The `grid` method is then called on `button1` to add it to `frame2`. The `row` and `column` parameters specify the position 
# of the button in the grid layout of the frame.
# The `bind` method is then called on `button1` to bind a function to the left mouse button click event (`"<Button-1>"`).
# In this case, the function bound to the event is `play`, which handles the player's turn in the game. When the player clicks
# on `button1`, the `play` function is called with an event object that contains information about the click.


button2 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30), bg="yellow" , relief=RAISED , borderwidth=5 )
button2.grid(row = 0 , column=1)
button2.bind("<Button-1>" , play)

button3 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30), bg="yellow" , relief=RAISED , borderwidth=5 )
button3.grid(row = 0 , column=2)
button3.bind("<Button-1>" , play)

#  second row 

button4 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30), bg="yellow" , relief=RAISED , borderwidth=5 )
button4.grid(row = 1 , column=0)
button4.bind("<Button-1>" , play)

button5 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30), bg="yellow" , relief=RAISED , borderwidth=5 )
button5.grid(row = 1 , column=1)
button5.bind("<Button-1>" , play)

button6 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30), bg="yellow" , relief=RAISED , borderwidth=5 )
button6.grid(row = 1 , column=2)
button6.bind("<Button-1>" , play)

#  third row 

button7 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30) , bg="yellow" , relief=RAISED , borderwidth=5)
button7.grid(row = 2 , column=0)
button7.bind("<Button-1>" , play)

button8 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30), bg="yellow" , relief=RAISED , borderwidth=5 )
button8.grid(row = 2 , column=1)
button8.bind("<Button-1>" , play)

button9 = Button(frame2 , text= " " , width=4 , height=2  , font=("Arial" , 30) , bg="yellow" , relief=RAISED , borderwidth=5)
button9.grid(row = 2 , column=2)
button9.bind("<Button-1>" , play)

restartButton = Button(frame2 , text="Restart Game" , width=19 , height=1 , font=("Arial" , 20) , bg="Green" , relief=RAISED , borderwidth=5 , command=restartGame )
restartButton.grid(row=4 , column=0 , columnspan=3)

buttons = [button1 , button2 , button3 , button4 , button5 , button6 , button7 , button8, button9]

root.mainloop()
```
